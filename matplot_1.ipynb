Line

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a data
overs = [1,2,3,4,5,6] # X
runs = np.cumsum([10, 2, 4, 14, 8, 12]) # y 
```

```python
plt.plot(overs,runs)
```

```python
# increase the line width and assining labes
plt.plot(overs,runs,linewidth=10,color='red',label='CSK')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('CSK runs for 1-6 overs')
plt.legend()
```

```python
rr_runs = np.cumsum([0,3,15,18,20,12])
rr_runs
```

```python
plt.plot(overs,runs,linewidth=10,color='yellow',label='CSK')
plt.plot(overs,rr_runs,linewidth=10,color='magenta',label='CSK')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('CSK runs for 1-6 overs')
plt.legend()
```
